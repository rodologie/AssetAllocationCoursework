# EQR Case Study - Qualitative Questions

## Question 1

Based on the evidence presented in the case study and the exhibits, AQR should not launch the Momentum Funds. While the long-only portfolio has an average annual return of 20.42% compared to the market return of 11% (Exhibit 11), the data reveals that the "long-only" retail product is a dangerously diluted version of the academic strategy, exposing investors to significantly higher risks without the downside protection that makes momentum attractive.

The most damning evidence against the launch is found in Exhibit 4, specifically regarding performance during market crises. AQR's marketing relies on the academic theory that momentum works as a hedge. For instance, in 2008, the classic "long-short" momentum strategy (High-Low) functioned perfectly, delivering a return of 30.37% while the market collapsed.

However, the retail fund cannot short; it can only hold the "High" portfolio (Decile 10). In that same year, the long-only "High" portfolio crashed by 38.84%. By stripping out the short positions, AQR removes the "crisis alpha." Retail investors expecting a diversifier will instead receive a product that amplifies their losses during market downturns.

Furthermore, Exhibit 3 proves that the retail product is fundamentally riskier than the academic strategy suggests. The academic long-short strategy is market-neutral with a beta of -0.08. In contrast, the long-only "High" portfolio has a beta of 1.16. This means the fund is not an alternative strategy but a "high-octane" equity product that is 16% more volatile than the broad market. The long-only portfolio has a standard deviation of 28.52%, which is significantly higher than the long-short strategy at 21.21%.

This excess volatility makes the fund unsuitable for "flighty" retail capital, as investors are likely to panic and sell during the inevitable sharp drawdowns shown in the data. While Exhibit 5 shows a low correlation between momentum (UMD) and the market of -0.03, this data is misleading for the retail launch. The correlation in Exhibit 5 is based on the long-short (UMD) factor. As noted in Exhibit 3, the long-only product has a beta of 1.16, implying a very high positive correlation to the market.

AQR would be selling a product based on the diversification benefits of UMD, but delivering a product that moves in lockstep with the market. When the market falls, this fund will fall faster, failing to provide the "value hedge" that is the primary argument for its existence.

## Question 2

Since SEC regulations prohibit mutual funds from marketing hypothetical "back-tested" returns, the index is essential. It serves as a compliant proxy to demonstrate the strategy's historical efficacy to retail investors who cannot see the hedge fundâ€™s internal data. Without it, AQR is selling a volatile "black box".

In addition, standard academic benchmarks (Fama-French UMD) are inappropriate because they are long-short and include illiquid small caps that a hedge fund cannot trade, especially for a retail investor audience. Also UMD is rebalanced monthly, which creates too must theoretical trading costs for a mutural fund. By launching a custom long-only, large-cap momentum index, AQR creates a realistic yardstick that matches the actual fund's constraints.

However, and this is clearly a double-edged sword, the index creates a strict expectation. If AQR optimises the fund to save taxes or trading costs (e.g. delaying a trade), the fund's performance will deviate from the index, resulting in "tracking error". Investors may view this deviation as a manager failure, even if it saves them money.

## Question 3

**Primary: AQR Large-Cap Momentum Index**

This is the most accurate measure of the strategy's specific universe (long-only, liquid US stocks).

**Will it beat it?** No, due to fees. While the active manager can add value by trading more smartly than the rigid quarterly index (e.g. rebalancing monthly), the "friction" of real-world investing makes beating the index difficult. The index pays no transaction costs, no bid-ask spreads, and no management fees. The fund must pay all three, as emphasised partly in the case study text. It is highly unlikely the fund will beat its own "paper" index after expenses.

**Secondary: Russell 1000 Growth**

Since momentum stocks are often high-growth "winners", this is the closest standard market benchmark for large-cap growth stocks (i.e. low B/M ratios). The case notes the strategy has historically outperformed this by 320bps.

**Will it beat it?** Yes, as is evident from the back-test stated in the case study.

**Broad Market: Russell 1000 / S&P 500**

This serves to measure the "beta" risk relative to the overall market.

**Will it beat it?** Historical data suggests the momentum factor generates significant excess returns (alpha). The academic "long" portfolio (Decile 10) averaged 20.42% annually versus the market's ~11%, suggesting it can absorb the fees and still outperform the broad market over a full cycle.
